```python
def tokenize(x):
    x = re.sub(r'http\S+', '', x)
    x = re.sub(r"#(\w+)", '', x)
    x = re.sub(r"@(\w+)", '', x)
    x = re.sub(r'[^\w\s]', '', x)
    x = x.strip().lower()
    x = word_tokenize(x)
    return x
```

```python
DATA['textt'] = DATA['text'].apply(lambda x: tokenize(x))
DATA=DATA.drop(['text'], axis=1)
```
